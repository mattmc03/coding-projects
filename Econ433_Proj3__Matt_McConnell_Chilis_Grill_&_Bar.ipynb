```python

# List of file names
files = ['weekly_patterns_2018_sample.csv.zip', 
         'weekly_patterns_2019_sample.csv.zip',
         'weekly_patterns_2020_sample.csv.zip', 
         'weekly_patterns_2021_sample.csv.zip',
         'weekly_patterns_2022_sample.csv.zip']


total_obs = 0
for file in files:
    # Read the data
    data = pd.read_csv(file, compression='zip')
    total_obs += len(data)
print(total_obs)
data.dtypes
chilis_obs = 0
chilis_df = pd.DataFrame() #initializing an empty df

for file in files:
    # Read the data
    data = pd.read_csv(file, compression='zip')
    
    # Filter the data for Chili's Grill & Bar
    data = data.loc[data['brands'].isin(["Chili's Grill & Bar"])]
    
    # Concatenate the data
    chilis_df = pd.concat([chilis_df, data], ignore_index = True)
    
    chilis_obs = len(chilis_df)
    
print(chilis_obs)
from datetime import datetime, date, timedelta
chilis_df['date_range_start'] = chilis_df['date_range_start'].str[:10]
chilis_df['date_range_end'] = chilis_df['date_range_end'].str[:10]
chilis_df.head(5)

chilis_df['year'] = chilis_df['date_range_start'].str[:4]
chilis_df['date_range_start'] = pd.to_datetime(chilis_df
                                               ['date_range_start'])
chilis_df['date_range_end'] =  pd.to_datetime(chilis_df
                                              ['date_range_end'])
# Split the variable visits_by_day into new variables
chilis_df[['dailyvisits0', 'dailyvisits1', 'dailyvisits2'
           , 'dailyvisits3',
    'dailyvisits4', 'dailyvisits5', 
           'dailyvisits6']] = chilis_df['visits_by_day'].str.split(',', expand=True)
chilis_df['dailyvisits0'] = chilis_df['dailyvisits0'].str[1:3]
chilis_df['dailyvisits6'] = chilis_df['dailyvisits6'].str[0:2]

chilis_df['dailyvisits0'] = chilis_df['dailyvisits0'].astype(float)
chilis_df['dailyvisits1'] = chilis_df['dailyvisits1'].astype(float)
chilis_df['dailyvisits2'] = chilis_df['dailyvisits2'].astype(float)
chilis_df['dailyvisits3'] = chilis_df['dailyvisits3'].astype(float)
chilis_df['dailyvisits4'] = chilis_df['dailyvisits4'].astype(float)
chilis_df['dailyvisits5'] = chilis_df['dailyvisits5'].astype(float)

chilis_df['dailyvisits6'] = chilis_df['dailyvisits6'].str.replace(']', '')

chilis_df['dailyvisits6'] = chilis_df['dailyvisits6'].astype(float)

chilis_df['id'] = chilis_df.index
chilis_df['index'] = chilis_df.index
long_chilis_df = pd.wide_to_long(chilis_df,
                                 ['dailyvisits'
                                 ], i = 'index', j = 'day', sep = '')
long_chilis_df['dailyvisits'] = long_chilis_df['dailyvisits'].astype(int)
agg = long_chilis_df.groupby(['year']).agg(
{'dailyvisits': 'mean' })

long_chilis_df = long_chilis_df.reset_index()

long_chilis_df['date'] = long_chilis_df['date_range_start'
                                       ] + pd.to_timedelta(long_chilis_df['day'],
                                                           unit='D')

long_chilis_df['dayoftheweek'] = long_chilis_df['date'].dt.day_name()
threshold = 34 #I chose this threshold because 25% of the data will be greater than it: 
                #this gets the top quartile

long_chilis_df['manyvisits'] = np.where(long_chilis_df['dailyvisits'] > threshold, 1, 0)
table = pd.crosstab(long_chilis_df['region'] , long_chilis_df['manyvisits'] )
tablepercent = pd.crosstab(long_chilis_df['region'] , long_chilis_df['manyvisits'] ,
                           normalize = 'index')*100
#same table as above but with percentages instead of total number counts
#Done to control for state size and see which states have best performing resturants
long_chilis_df['core_biz_area'] = np.where((long_chilis_df['region'] == 'AL')
                                           | (long_chilis_df['region'] =='AR')
                                           | (long_chilis_df['region'] =='FL') 
                                           | (long_chilis_df['region'] =='GA')
                                           |(long_chilis_df['region'] == 'KS') 
                                           | (long_chilis_df['region'] =='LA')
                                           |(long_chilis_df['region'] == 'MI') 
                                           |(long_chilis_df['region'] == 'MO') 
                                           |(long_chilis_df['region'] == 'MS') 
                                           | (long_chilis_df['region'] =='NC') 
                                           |(long_chilis_df['region'] =='OK') 
                                           |(long_chilis_df['region'] == 'TX') 
                                           |(long_chilis_df['region'] == 'WV'), 1, 0)

long_chilis_df.head()

core_agg = long_chilis_df.groupby(['core_biz_area']).agg(
{'dailyvisits': 'mean' })

core_agg.head()
```

```python
MergedSummary['Post03132020'] = (MergedSummary[
    'date'] >= '2020-03-13').astype(int)
```

```python
blue_states = ['ME', 'NH', 'MA', 'CT', 'NY', 'RI', 'NJ', 
               'MD', 'VA', 'IL', 'MN', 'CO', 'NM', 'NV', 
               'CA', 'OR', 'WA']
MergedSummary['BlueState'] = MergedSummary['state'].isin(blue_states).astype(int)
```

```python
MergedSummary['BluePost'] = MergedSummary['BlueState'] * MergedSummary['Post03132020']
MergedSummary.head()
```

```python
import statsmodels.formula.api as smf
```

```python
def log_plus_1(x):
    return np.log(x + 1)
```

```python
results = smf.ols('log_plus_1(SumDailyVisits_y) ~ C(Post03132020) + C(BlueState) + C(BluePost)',
                  data=MergedSummary).fit()


results.summary()
```

The coefficients on Post03132020 & BlueState are both statistically significant at the 95% level as seen by the P values being 0. The coefficient on BluePost however is not significant at the 95 % level with a P value of almost .6, & 0 being included in the 95% confidence Interval

Since we are in the log-normal form, the three coefficients mean that: 1.) being after the national emergency decreased the sum of visits per day by 9.5%. 2.) Being a blue state decreases the sum of visits perday by 48.5% & 3.) being a blue state post pandemic has an expected 1.2% decrease in sum daily visits. 

Thus I would accept the null hypothesis that The national emergency announced on March 13, 2020 does not make a difference on average SumDailyVisits of your own chain between red and blue states”? 

The national emergency itself definitely had an effect, but Blue and red states saw similar percentage decreases in dailyvisits

```python
from linearmodels import PanelOLS
```

```python
df2 = MergedSummary.copy()
df2 = MergedSummary.set_index(['state', 'date'])
df2.head()
```

```python
df2[['BlueState', 'Post03132020', 'BluePost']].head()
```

```python
mod2 = PanelOLS(log_plus_1(df2['SumDailyVisits_y']), 
               df2[['BlueState', 'Post03132020', 'BluePost']], 
               entity_effects = True,
               drop_absorbed = True)
results2 = mod2.fit(cov_type='clustered')
print(results2)
```

The above code is for just State FE. Thus BlueState gets absorbed. Interestingly, now BluePost has a statistically significant coefficient suggesting that blue states did lead to a decrease of 7.47% of avg daily visits. 

```python
mod = PanelOLS(log_plus_1(df2['SumDailyVisits_y']), 
               df2[['BlueState', 'Post03132020', 'BluePost']], 
               time_effects = True,
               drop_absorbed = True)
results = mod.fit(cov_type='clustered')
print(results)
```

Just including the time FE Guves a super high coefficient for BlueState, but this is because as seen in the previous data, chilis core business area is mostly red states like Texas, Alabama, and Florida. so without including state FE this number is inflated

```python
mod = PanelOLS(log_plus_1(df2['SumDailyVisits_y']), 
               df2[['BlueState', 'Post03132020', 'BluePost']], 
               time_effects = True, entity_effects = True,
               drop_absorbed = True)
results = mod.fit(cov_type='clustered')
print(results)
```

Including both effects leaves us with just the BluePost variable, and the coefficient is statistically significant suggesting differently that the original regression from part a that being a bluestate post 03132020 does in fact decrease average daily visits by 9.5 percent

Including both Fixed effects is definately better for hypothesis testing as it allows us to see through state by state differences and get a better, and more importantly statistically significant coefficient. Thus using the Fixed effects we can reject the aforementioned hypothesis whereas without adding the fixed effects we could not reject. 

```python
difference = log_plus_1(MergedSummary['SumDailyVisits_y']) - log_plus_1(MergedSummary['SumDailyVisits_x'])

results2 = smf.ols('difference ~ C(Post03132020) + C(BlueState) + C(BluePost)',
                   data=MergedSummary).fit()


results2.summary()
```

The Coefficients on Post03132020 and BlueState are still very similar to the previous regression. However, BluePost now has a strong positive Coefficient. This suggests that Chilis in blue states greatly outperformed applebees in blue states after the pandemic.
